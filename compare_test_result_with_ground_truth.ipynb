```python
from sklearn.metrics import confusion_matrix
import pandas as pd

cm = confusion_matrix(y_test, y_pred)

results = []

for i in range(len(cm)):

    TP = cm[i, i]

    FP = cm[:, i].sum() - TP

    FN = cm[i, :].sum() - TP

    TN = cm.sum() - TP - FP - FN

    results.append([
        i,
        TP,
        FP,
        FN,
        TN
    ])

tp_fp_fn_df = pd.DataFrame(
    results,
    columns=[
        "Class",
        "TP",
        "FP",
        "FN",
        "TN"
    ]
)

print(tp_fp_fn_df)
```
