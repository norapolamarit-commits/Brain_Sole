```python

def simulate_group(n, risk_level):
    if risk_level == 0:   # Low Risk
        data = {
            "fsr_forefoot_left": np.random.normal(55, 12, n),
            "fsr_forefoot_right": np.random.normal(55, 12, n),
            "fsr_midfoot_left": np.random.normal(35, 9, n),
            "fsr_midfoot_right": np.random.normal(35, 9, n),
            "fsr_heel_left": np.random.normal(45, 10, n),
            "fsr_heel_right": np.random.normal(45, 10, n),
            "speed_kmh": np.random.normal(8.5, 1.5, n),
            "cadence_spm": np.random.normal(165, 12, n),
            "stride_length_m": np.random.normal(1.10, 0.15, n),
            "ground_contact_time_ms": np.random.normal(250, 30, n),
        }

    elif risk_level == 1: # Medium Risk
        data = {
            "fsr_forefoot_left": np.random.normal(65, 15, n),
            "fsr_forefoot_right": np.random.normal(48, 15, n),
            "fsr_midfoot_left": np.random.normal(45, 12, n),
            "fsr_midfoot_right": np.random.normal(35, 12, n),
            "fsr_heel_left": np.random.normal(58, 14, n),
            "fsr_heel_right": np.random.normal(42, 14, n),
            "speed_kmh": np.random.normal(10.5, 2.0, n),
            "cadence_spm": np.random.normal(150, 14, n),
            "stride_length_m": np.random.normal(1.25, 0.18, n),
            "ground_contact_time_ms": np.random.normal(290, 35, n),
        }

    else:                 # High Risk
        data = {
            "fsr_forefoot_left": np.random.normal(85, 18, n),
            "fsr_forefoot_right": np.random.normal(35, 15, n),
            "fsr_midfoot_left": np.random.normal(70, 15, n),
            "fsr_midfoot_right": np.random.normal(30, 12, n),
            "fsr_heel_left": np.random.normal(80, 18, n),
            "fsr_heel_right": np.random.normal(30, 12, n),
            "speed_kmh": np.random.normal(13.0, 2.0, n),
            "cadence_spm": np.random.normal(138, 12, n),
            "stride_length_m": np.random.normal(1.40, 0.18, n),
            "ground_contact_time_ms": np.random.normal(335, 35, n),
        }

    df = pd.DataFrame(data)

    for col in df.columns:
        df[col] = df[col].clip(lower=0)

    df["risk_level"] = risk_level
    return df


def add_features(df):
    df["left_pressure"] = (
        df["fsr_forefoot_left"] +
        df["fsr_midfoot_left"] +
        df["fsr_heel_left"]
    )

    df["right_pressure"] = (
        df["fsr_forefoot_right"] +
        df["fsr_midfoot_right"] +
        df["fsr_heel_right"]
    )

    df["pressure_imbalance"] = abs(
        df["left_pressure"] - df["right_pressure"]
    )

    df["total_pressure"] = (
        df["left_pressure"] + df["right_pressure"]
    )

    df["forefoot_pressure"] = (
        df["fsr_forefoot_left"] +
        df["fsr_forefoot_right"]
    )

    df["midfoot_pressure"] = (
        df["fsr_midfoot_left"] +
        df["fsr_midfoot_right"]
    )

    df["heel_pressure"] = (
        df["fsr_heel_left"] +
        df["fsr_heel_right"]
    )

    df["forefoot_heel_ratio"] = (
        df["forefoot_pressure"] / (df["heel_pressure"] + 1)
    )

    df["left_right_ratio"] = (
        df["left_pressure"] / (df["right_pressure"] + 1)
    )

    return df


def create_dataset():
    low = simulate_group(2000, 0)
    medium = simulate_group(1750, 1)
    high = simulate_group(1250, 2)

    df = pd.concat([low, medium, high], ignore_index=True)
    df = add_features(df)

    df = df.sample(frac=1, random_state=42).reset_index(drop=True)

    return df


np.random.seed(42)

df = create_dataset()

print("Dataset distribution:")
print(df["risk_level"].value_counts())
```
